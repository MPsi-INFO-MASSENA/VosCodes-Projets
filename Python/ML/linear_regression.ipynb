```python
import numpy as np 
from matplotlib import pyplot as plt 
plt.style.use('seaborn')
```

```python
trainset = np.array([[1,1,0.5],[1,2,1.5],[1,3,3.5]])
X0 = trainset[:,:2]
Y0 = trainset[:,2].reshape(3,1)
lr = 0.05
theta = np.zeros((2,1))

h = lambda x : np.dot(X0,theta)
```

```python
for _ in range(500):
    #vectorization pour accelerer le calcul et minimiser le nombre de ligne de code (ie utilisation de : matrice bloc et assimilation de (a) à a )
    theta = theta - (lr/3)*np.dot(X0.T, h(X0)-Y0) 
    
```

```python
x = np.arange(0,4,0.25)
y = theta[0] + x*theta[1]
```

```python
plt.plot(x,y)
plt.scatter(X0[:,1],Y0,c ="red")
```
